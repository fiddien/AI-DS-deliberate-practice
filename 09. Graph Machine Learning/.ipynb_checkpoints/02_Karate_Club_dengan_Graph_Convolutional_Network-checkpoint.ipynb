# NetworkX

NetworkX adalah package yang paling sering digunakan untuk membuat, memanipulasi, dan mengeksplorasi tipe data graf.

````python
# Import NetworkX package
import networkx as nx

# Cek versi package
print(nx.__version__)
````

## Graf

Graf di NetworkX dapat didefinisikan dari awal, atau mengubah tipe data lain.

Ada beberapa [tipe graf di NetworkX](https://networkx.org/documentation/stable/reference/classes/index.html). Selengkapnya silakan baca di website dokumentasi NetworkX.

````python
# Membuat graf tak berarah G
G = nx.Graph()
print(G.is_directed())

# Membuat graf tak berarah H
H = nx.DiGraph()
print(H.is_directed())

# Menambah atribut pada level graf
G.graph["Nama"] = "Tes" # nama graf ini adalah "Tes"
print(G.graph)
````

## Simpul

Kita dapat menambahkan simpul beratribut satu per satu atau sekaligus.

````python
# Menambah satu simpul `0` dengan atribut di level simpul
G.add_node(0, feature=5, label=0)

# Memanggil simpul `0`
node_0_attr = G.nodes[0]
print("Simpul 0 punya atribut {}".format(node_0_attr))

# Menambah banyak simpul dengan atributnya
# input: list dari (simpul, kamus-atribut)
G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
]) 

# Mengecek ke semua simpul
print("Simpul graf G:")
for n in G.nodes(data=True):
  print(n)

# Mendapatkan banyaknya simpul
num_nodes = G.number_of_nodes() # sama dengan G.order()
print("G punya {} simpul".format(num_nodes))
````

## Sisi

Sama seperti simpul, kita dapat menambahkan sisi beratribut satu per satu atau sekaligus.

````python
# Menambah sebuah sisi dengan beban sisi 0.5
G.add_edge(0, 1, weight=0.5)   

# Melihat atribut sisi (0, 1)
edge_0_1_attr = G.edges[(0, 1)]
print("Sisi (0, 1) punya atribut {}".format(edge_0_1_attr))

# Menambah banyak sisi dengan atributnya
# input: list dari (sisi, kamus-atribut)
G.add_edges_from([
  (1, 2, {"weight": 0.3}),
  (2, 0, {"weight": 0.1})
])

# Mengecek ke semua simpul
for edge in G.edges():
  print(edge)

# Mendapatkan banyaknya sisi
num_edges = G.number_of_edges() # sama dengan G.size()
print("G punya {} sisi".format(num_edges))
````

## Visualisasi

NetworkX dapat memberikan visualisasi statis dari graf yang dibangun di atas modul `matplotlib.pyplot`.

Ada banyak cara menempatkan simpul-simpul pada graf, menggunakan berbagai algoritma. Baca dokumentasinya [di sini](https://networkx.org/documentation/stable/reference/drawing.html).

````python
# Menggambar graf
nx.draw(G, with_labels=True)
````

````python
# menambah variasi syling pada visualisasi
node_colors = [G.nodes[n]['feature'] for n in G.nodes]
edge_widths = [G.edges[e]['weight']*2 for e in G.edges]
node_sizes = [(G.nodes[n]['label']+1)*300 for n in G.nodes]

# merekam posisi graf G dengan layout algoritma spring
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color=node_colors, width=edge_widths, 
        node_size=node_sizes, cmap='Set2')
````

## Fungsi pada graf
NetworkX dapat mengkalkulasi berbagai properti umum suatu graf, seperti derajat simpul dan tetangga simpul.

````python
node_id = 1

# Derajat dari simpul 1
print("Simpul {} punya derajat {}".format(node_id, G.degree[node_id]))

# Dapatkan simpul-simpul tetangga 1
for neighbor in G.neighbors(node_id):
    print("Simpul {} punya tetangga {}".format(node_id, neighbor))
````

### Fungsi lainnya
Salah satu cara mengeksplorasi suatu graf adalah melihat [PageRank](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.link_analysis.pagerank_alg.pagerank.html#networkx.algorithms.link_analysis.pagerank_alg.pagerank) dari simpul.

````python
num_nodes = 4

# Membuat graf lintasan/garis dan mengubahnya jadi berarah
G = nx.DiGraph(nx.path_graph(num_nodes))

# Dapatkan pagerank untuk graf G
pr = nx.pagerank(G, alpha=0.8)
print(pr)

nx.draw(G, with_labels = True, node_size=[i*2000 for i in pr.values()])
````

## Mengimplementasikan Graph Neural Networks (GNNs)

Kita akan menggunakan salah satu operator GNN paling sederhana, yaitu **Graph Convolutional Network (GCN)** ([Kipf et al. (2017)](https://arxiv.org/abs/1609.02907)).

PyG menerapkannya ke [`GCNConv`](https://pytorch-geometric.readthedocs.io/en/latest/modules/nn.html#torch_geometric.nn.conv.GCNConv) yang dapat dipanggil dengan memasukan representsi fitur simpul, yaitu `x`, dan representasi kontektivitas graf, yaitu COO `edge_index` tadi.

### Keluaran dari GCN

Tujuan dari GNN adalah menerima input graf $G = (\mathcal{V}, \mathcal{E})$ dengan setiap simpul $v_i \in \mathcal{V}$ punya vektor fitur $X_i^{(0)}$, lalu mengeluarkan suatu *embedding* simpul $\mathcal{R}^d$. Dengan kata lain, kita ingin model mempelajari $f \to \mathcal{V} \times \mathcal{R}^d$, pemetaaan yang menghubungan setiap simpul kepada suatu representasi baru yang berguna untuk tugas kita. "Tugas" ini bisa beragam, mulai dari tugas pada level simpul, level sisi, atau regresi/klasifikasi level graf.

> Pada masalah Klub Karate, kita ingin mempelajari *embedding* yang berguna untuk mengklasifikasikan setiap simpul ke suatu komunitas.

Mari! Kita sudah siap untuk membuat Graph Neural Network pertama kita dengan mendefinisikan arsitektur modelnya di atas kelas`torch.nn.Module`.

````python
import torch
from torch.nn import Linear
from torch_geometric.nn import GCNConv

class GCN(torch.nn.Module):
    def __init__(self):
        super(GCN, self).__init__()
        torch.manual_seed(123)
        # objek 'dataset' sudah didefinisikan sebelumnya
        self.conv1 = GCNConv(dataset.num_features, 4)
        self.conv2 = GCNConv(4, 4)
        self.conv3 = GCNConv(4, 2)
        self.classifier = Linear(2, dataset.num_classes)

    def forward(self, x, edge_index):
        h = self.conv1(x, edge_index)
        h = h.tanh()
        h = self.conv2(h, edge_index)
        h = h.tanh()
        h = self.conv3(h, edge_index)
        embeddings = h.tanh()
        out = self.classifier(embeddings)

        return out, embeddings

model = GCN()
print(model)
````

Pada kode di atas, kita mendeskripsikan setiap batu bata model di fungsi `__init__` dan menjelaskan alur komputasinya di `forward`. Kita memulai dengan membuat tiga lapisan *graph convolution*. Setiap satu lapisan bertanggungjawab untuk menyambungkan informasi dari setiap tetangga 1 langkah (tetangga langsung/1-hop neighborhood). Namun, dengan adanya tiga lapisan, kita bisa menghubungkan informasi untuk tetangga 1 hingga 3 langkah sekaligus (3-hop neighborhood).

Selain itu, lapisan `GCNConv` mengurasi dimensi dari vektor fitur simpul dari 34 menjadi 2, *i.e.*, $34 \rightarrow 4 \rightarrow 4 \rightarrow 2$. Setiap `GCNConv` diakhiri dengan ketidaklinearan fungsi `tanh`.

Terakhir, kita melakukan transformasi linear biasa dengan ([`torch.nn.Linear`](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html?highlight=linear#torch.nn.Linear)) yang berperan menjadi pengklasifikasi yang menentukan simpul salah satu dari empat kelas (komunitas klub karate).

Kita mengeluarkan hasil klasifikasi sekalian dengan *embedding* simpul yang telah dilatih oleh GNN.

Kita coba melakukan satu kali *forward pass*.

````python
_, h = model(data.x, data.edge_index)
print(f'Ukuran embedding: {list(h.shape)}')

print("Embedding prediksi")
visualize(h, color=data.y)

print("Komunitas aslinya")
visualize(G, color=data.y)
````

Menariknya, bahkan sebelum kita melatihnya, model menghasilkan embedding simpul yang lumayan mirip dengan struktur komunitas aslinya. Padahal, kita menginisialisasi *weight* model secara random dan kita belum melakukan latihan apapun! Fenomena ini menunjukkan bahwa GNN mengenalkan kita pada *inductive bias* yang kuat, yaitu keterhubungan antara simpul, yang akhirnya menghasilkan *embedding* yang mirip untuk simpul yang saling berdekatan.

## Melatih model pada Jaringan Klub Karate

Bisakah kita melakukan latihan model dengan lebih baik?



Karena setiap *weight* di model kita *differentiable*, kita bisa menambah label lagi lalu latih modelnya dan lihat bagaimana pengaruhnya pada *embedding*. Di sini kita menggunakan teknik *semi-supervised* dan *transductive learning*: kita melatihan model hanya satu simpul untuk tiap kelas, tapi bisa lanjut menggunakan informasi dari graf inputnya.

Kita akan melatih model seperti melatih model PyTorch lainnya. Setelah mendefinisikan arsitektur jaringan model, kita deklarasikan *loss criterion* (di sini [`CrossEntropyLoss`](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html)) dan menggunakan optimizer gradient stokastik (di sini [`Adam`](https://pytorch.org/docs/stable/optim.html?highlight=adam#torch.optim.Adam)). Dalam satu iterasi epoch, kita mulai dengan membersihkan gradient, kemudian melakukan sekali forward pass, lalu menghitung loss yang dilanjut dengan mencari turunan gradient dan mengoptimisasi parameter model. Jika kamu belum familiar dengan skema ini, silakan cek [dokumentasi yang disediakan PyTorch](https://pytorch.org/tutorials/beginner/blitz/cifar10_tutorial.html#define-a-loss-function-and-optimizer).

Skenario *semi-supervised learning* kita diwujudkan dengan adanya baris:

```python 
loss = criterion(out[data.train_mask], data.y[data.train_mask])
```

Ketika menghitung *embedding*, kita memerhatikan seluruh simpul, namun dalam menghitung *loss* yang akan dioptimisasi, kita hanya melihat pada simpul latihan yang ada di `data.train_mask`. Perhatikan bahwa `out` adalah hasil prediksi komunitas simpul dan `data.y` adalah *ground-truth*-nya.

Mari kita jalankan kodenya dan lihat bagaimana *embedding* kita berubah seiring bergantinya epoch.

````python
import time
from IPython.display import Javascript
display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 460})'''))

model = GCN()
criterion = torch.nn.CrossEntropyLoss()  # kriteria loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  # optimizer

def train(data):
    optimizer.zero_grad()  # mengosongkan gradient
    out, h = model(data.x, data.edge_index)  # forward pass.

    loss = criterion(out[data.train_mask], data.y[data.train_mask])  # hitung loss
    loss.backward()  # menurunkan gradient
    optimizer.step()  # update parameter berdasarkan turunan gradient

    # Menghitung akurasi latihan pada empat simpul
    accuracy = {}
      # ambil indeks dengan nilai yang terbesar [0.6, 0.2, 0.7, 0.1] -> 2
    predicted_classes = torch.argmax(out[data.train_mask], axis=1) 
    target_classes = data.y[data.train_mask]
    accuracy['train'] = torch.mean(
        torch.where(predicted_classes == target_classes, 1, 0).float())
    
    # Menghitung akurasi validasi pada graf keseluruhan
    predicted_classes = torch.argmax(out, axis=1)
    target_classes = data.y
    accuracy['val'] = torch.mean(
        torch.where(predicted_classes == target_classes, 1, 0).float())

    return loss, h, accuracy, predicted_classes

for epoch in range(500):
    loss, h, accuracy, pred_class = train(data)
    # visualisasi embedding tiap 10 epoch
    if (epoch+1) % 10 == 0:
        visualize(h, color=data.y, epoch=epoch, loss=loss, accuracy=accuracy)
        time.sleep(0.3)
````

## Kesimpulan

````python
print("Simpul Klub Pred")
for i in range(data.num_nodes):
    print(f"{i:4} {data.y[i]:5} {pred_class[i]:4}  {'X' if data.y[i]!=pred_class[i] else ''}")
````

Seperti yang bisa dilihat, GCN 3 lapis kita bisa memisahkan komunitas dengan cukup baik dan mengklasifikasikan simpul-simpul dengan akurat.
