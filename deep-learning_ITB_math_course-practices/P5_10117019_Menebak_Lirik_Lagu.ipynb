Akan diambil lirik lagu-lagu Rhoma Irama sebagai data yang akan diolah. Kumpulan lirik bersumber dari `musiklib.org`.

Pertama, ambil HTML dari halaman yang mengandung seluruh link menuju tiap halaman lirik lagu Rhoma Irama.

```python
import requests

url = "https://musiklib.org/rhoma_irama-lirik_lagu.htm"
res = requests.get(url)
html_page = res.content

from bs4 import BeautifulSoup

soup = BeautifulSoup(html_page, 'html.parser')
text = soup.find_all(text=True)
set([t.parent.name for t in text])
```

Semua link berada pada elemen `href` di bawah parent `a`.

```python
links = []
soup = BeautifulSoup(html_page)
for link in soup.findAll('a'):
    links.append(link.get('href'))
print(links)
```

Filter, ambil link yang merupakan link menuju halaman lirik lagu Rhoma Irama.

```python
rhoma_links = ['https://musiklib.org/'+l for l in links if l is not None and '/rhoma' in l][2:]
rhoma_links[:5]
```

## 1.2 Mengimpor teks dari halaman web HTML
Sama seperti tahap di atas. Kali ini teks lirik lagu berada di bawah parent `div` dengan sedikit pemenggalan di akhir string.

```python
def get_lyrics(url):
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    text = soup.find_all(text=True)
    set([t.parent.name for t in text])
    output = ''
    whitelist = ['div']

    for t in text:
        # print(t.parent.name)
        # if t.parent.name not in blacklist:
        if t.parent.name in whitelist:
            output += '{} '.format(t)
    
    return output[:-312]
```

~~Mari kita gunakan 100 lirik lagu saja dulu.~~
Tiap elemen di `lyrics` adalah *string* untuk lirik satu lagu.

```python
data = []
for song in rhoma_links:
    data.append(get_lyrics(song))
data[:5]
```

```python
import re

def cleaning(string):
    ''' cleaning a string, only leaves alphabetic character '''
    return re.sub("[^a-zA-Z]+", " ", string).lower().split()

def flatten(list_of_list):
    ''' flatten any list of list '''
    return [item for sublist in list_of_list for item in sublist]
```
